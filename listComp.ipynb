```python
transposed = []
matrix = [[1, 2, 3, 4], [5, 6, 7, 8]]
for col in range(len(matrix[0])):
    trans_row = []
    for row in matrix:
        trans_row.append(row[col])
    transposed.append(trans_row)
print(transposed)
    
```

```python
matrix = [[1, 2, 3, 4], [5, 6, 7, 8]]
transposed = [[ row[col] for row in matrix]
              for col in range(len(matrix[0]))]
print(transposed)
```

```python
matrix = [[1, 5], [2, 6], [3, 7], [4, 8]]
transposed = [[ row[col] for row in matrix]
              for col in range(len(matrix[0]))]
print(transposed)
```
